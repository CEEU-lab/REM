```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

# 1. Precios de suelo

```python
sim_terrenos = gpd.read_file("../runs/properaty_sim_prices_terrenos.geojson")

f,ax = plt.subplots(figsize=(12,8))
sim_terrenos.plot(ax=ax, column='usdm2', k=5, cmap='plasma', 
                  scheme='NaturalBreaks', edgecolor='gray', linewidth=0.05,
                  legend=True, legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)})
ax.set_axis_off()
ax.set_title('USD/M2-Terrenos simulado en base a Properaty train',fontsize=12)
plt.tight_layout();
```
